```python
import pandas as pd
from goose3 import Goose
import os

# Create an instance of Goose
goose = Goose()

# Define the output directory
output_dir = 'extracted_articles'
os.makedirs(output_dir, exist_ok=True)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Fetch the article using Goose
    article = goose.extract(url=url)
    
    # Get the article title and text
    title = article.title
    text = article.cleaned_text
    
    # Save the article title and text to a text file
    filename = f"{url_id}.txt"
    filepath = os.path.join(output_dir, filename)
    
    with open(filepath, 'w', encoding='utf-8') as file:
        file.write(f"Title: {title}\n\n")
        file.write(f"Article:\n{text}")

    print(f"Article saved: {filename}")

print("Article extraction completed.")
```

```python
import os
from textblob import TextBlob
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
import syllapy
```

```python

# Download NLTK data files
nltk.download('punkt')
nltk.download('stopwords')

# Function to compute textual analysis variables
def analyze_text(text):
    # Tokenize text
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    
    # Counts
    word_count = len(words)
    sentence_count = len(sentences)
    
    # Calculate Average Word Length
    avg_word_length = sum(len(word) for word in words) / word_count if word_count else 0
    
    # Calculate Average Sentence Length
    avg_sentence_length = word_count / sentence_count if sentence_count else 0
    
    # Calculate Complex Word Count and Percentage of Complex Words
    complex_words = [word for word in words if syllapy.count(word) > 2]
    complex_word_count = len(complex_words)
    percentage_complex_words = (complex_word_count / word_count) * 100 if word_count else 0
    
    # Calculate FOG Index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words) if sentence_count else 0
    
    # Calculate Syllables per Word
    syllables_per_word = sum(syllapy.count(word) for word in words) / word_count if word_count else 0
    
    # Personal Pronouns Count
    personal_pronouns = sum(1 for word in words if word.lower() in {'i', 'me', 'my', 'mine', 'myself', 'we', 'us', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves'})
    
    # TextBlob Sentiment Analysis
    blob = TextBlob(text)
    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity
    
    return {
        'POSITIVE SCORE': max(polarity_score, 0),
        'NEGATIVE SCORE': min(polarity_score, 0),
        'POLARITY SCORE': polarity_score,
        'SUBJECTIVITY SCORE': subjectivity_score,
        'AVG SENTENCE LENGTH': avg_sentence_length,
        'PERCENTAGE OF COMPLEX WORDS': percentage_complex_words,
        'FOG INDEX': fog_index,
        'AVG NUMBER OF WORDS PER SENTENCE': avg_sentence_length,
        'COMPLEX WORD COUNT': complex_word_count,
        'WORD COUNT': word_count,
        'SYLLABLE PER WORD': syllables_per_word,
        'PERSONAL PRONOUNS': personal_pronouns,
        'AVG WORD LENGTH': avg_word_length
    }

# Load the output structure file
output_structure_file = 'Output Data Structure.xlsx'
output_structure_df = pd.read_excel(output_structure_file)
output_columns = output_structure_df.columns.tolist()

# Prepare to store results
results = []

# Directory containing extracted articles
extracted_articles_dir = 'extracted_articles'

# Iterate through the extracted article files
for filename in os.listdir(extracted_articles_dir):
    if filename.endswith('.txt'):
        url_id = filename.replace('.txt', '')
        url = output_structure_df[output_structure_df['URL_ID'] == url_id]['URL'].values[0]
        filepath = os.path.join(extracted_articles_dir, filename)
        
        # Read the article text
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read()
            # Extract article text after the title
            _, text = content.split('\n\n', 1)  # Splitting title and text
        
        # Analyze the text
        analysis_results = analyze_text(text)
        analysis_results['URL_ID'] = url_id
        analysis_results['URL'] = url
        
        # Append results
        results.append(analysis_results)

# Create DataFrame from results
results_df = pd.DataFrame(results)

# Reorder columns according to the output structure file
results_df = results_df[output_columns]
```
